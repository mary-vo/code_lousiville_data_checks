```python
import pandas as pd
```

```python
# read file
lou_inspection = pd.read_excel("louisville_inspection.xlsx")
lou_inspection
```

```python
# Change all headers to lower case letters
lou_inspection.columns = lou_inspection.columns.str.lower()
lou_inspection
```

```python
# Force all string values to be lowercase
lou_inspection = lou_inspection.applymap(lambda x: x.lower() if type(x) == str else x)
lou_inspection
```
